Crear una función que reciba dos letras (parámetros) y que verifique si son iguales:

Si son iguales, imprimir "Las letras son iguales".,

De lo contrario, imprimir "Las letras son diferentes".

```python
#Repaso
#Challenge 1:
letra1 = input("Ingrese una letra")
letra2 = input("Ingrese una segunda letra")

def verificar_letra(l1,l2):
    if l1 == l2:
        print("Las letras son iguales")
    else:
        print("Las letras no son iguales")

verificar_letra(letra1,letra2)
```

Pista: cantidad de letras de una palabra en python

```python
#challenge 2: 
#paso 1: definir parametros
parametro_1 = input("Ingrese una palabra")
parametro_2 = int(input("Ingrese un numero"))

def comparacion_de_parametros(palabra, numero):
  cantidad_de_letras = len(palabra) #paso 2: determinar cantidad de letras
  if cantidad_de_letras == numero:  #paso 3: comparar párametros
    return True
  else:
    return False


comparacion_de_parametros(parametro_1, parametro_2)
```

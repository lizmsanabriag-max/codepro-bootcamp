Crear dos metodos para nuestras mascotas y ejecutar, debe ser algo especifico de esa mascota.
Ejemplo: "Cuando hablo, digo X" "Cuando me muevo, hago X"

```python
#paso 1: definir los atributos de clase
class animales:
    especie = "felino"
    #paso 2: definir los atributos de instancia
    def __init__(self, nombre, color):
            self.nombre = nombre
            self.color = color

    def hablar(self, maullo):
        print(f"Cuando hablo digo {maullo}")

    def mover(self,movimiento):
        print(f"Cuando me muevo, hago {movimiento}")

gato1 = animales("Kuro", "negro")
print(f"Hola soy {gato1.nombre}, soy un gatito de color {gato1.color}")
gato1.hablar("Miau")
gato1.mover("Tap tap")
```
